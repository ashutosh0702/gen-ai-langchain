```python
from langchain_openai import ChatOpenAI
import os
```

```python
os.getenv("OPEN_API_KEY")



model = ChatOpenAI()
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages = [
    SystemMessage(content="Translate the following from English into Italian"),
    HumanMessage(content="hi!"),
]

model.invoke(messages)
```

```python
from langchain_huggingface import HuggingFaceEndpoint

import os
os.getenv("HUGGINGFACEHUB_API_KEY")
```

```python
llm_huggingface = HuggingFaceEndpoint(repo_id="google/flan-t5-base")

llm_huggingface.invoke("What is the capital of India ?")
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
```

```python
prompt_template = PromptTemplate(input_variables=["country"], template="Tell me the capital of this {country}")
```

```python
chain = LLMChain(llm=llm_huggingface, prompt=prompt_template)

chain.invoke("India")
```
